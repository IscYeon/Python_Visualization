```python
df = pd.read_csv('Health_2021.csv')
df.head()
```

```python
df.SIDO.value_counts().index.sort_values()
# 11 서울특별시 42 강원도 26 부산광역시 43 충청북도 27 대구광역시 44 충청남도 28 인천광역시 45 전라북도 
# 29 광주광역시 46 전라남도 30 대전광역시 47 경상북도 31 울산광역시 48 경상남도 
# 36 세종특별자치시 49 제주특별자치도 41 경기도
```

```python
df = df.assign(CITY=df.SIDO)
df = df.assign(Country='대한민국')
# df
```

```python
df['CITY'] = df['CITY'].replace(to_replace=11, value='서울특별시')
df['CITY'] = df['CITY'].replace(to_replace=26, value='부산광역시')
df['CITY'] = df['CITY'].replace(to_replace=27, value='대구광역시')
df['CITY'] = df['CITY'].replace(to_replace=28, value='인천광역시')
df['CITY'] = df['CITY'].replace(to_replace=29, value='광주광역시')
df['CITY'] = df['CITY'].replace(to_replace=30, value='대전광역시')
df['CITY'] = df['CITY'].replace(to_replace=31, value='울산광역시')
df['CITY'] = df['CITY'].replace(to_replace=36, value='세종특별자치시')
df['CITY'] = df['CITY'].replace(to_replace=41, value='경기도')
df['CITY'] = df['CITY'].replace(to_replace=42, value='강원도')
df['CITY'] = df['CITY'].replace(to_replace=43, value='충청북도')
df['CITY'] = df['CITY'].replace(to_replace=44, value='충청남도')
df['CITY'] = df['CITY'].replace(to_replace=45, value='전라북도')
df['CITY'] = df['CITY'].replace(to_replace=46, value='전라남도')
df['CITY'] = df['CITY'].replace(to_replace=47, value='경상북도')
df['CITY'] = df['CITY'].replace(to_replace=48, value='경상남도')
df['CITY'] = df['CITY'].replace(to_replace=49, value='제주특별자치도')

# df
```

```python
print(df['SIDO'].value_counts())
print(df['CITY'].value_counts())
```

```python
df.to_csv("Health_Korea.csv", index = False)
```
